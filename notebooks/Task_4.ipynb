По результатам экспериментов установлено, что наилучшее качество демонстрирует модель LeNet-5, её максимальная точность на валидационной выборке составила около 0.914–0.915, а точность на тестовой выборке — 0.9575. Модель со сверточными слоями (первое задание) достигла максимальной точности на валидации около 0.904, при этом при увеличении числа эпох наблюдалось ухудшение качества на валидации, что указывает на начало переобучения. Модель с пулингом (второе задание) показала наихудшие результаты, её точность на валидационной выборке стабилизировалась около 0.77–0.78, а на тестовой выборке составила около 0.83, практически не улучшаясь при увеличении числа эпох.

Полученные различия объясняются архитектурными особенностями моделей. Во втором задании пулинг-слои не содержат обучаемых параметров и выполняют только операцию уменьшения размерности, не обеспечивая извлечение признаков изображений. В результате модель обладает недостаточной выразительной способностью и демонстрирует недообучение. В первом задании использование сверточных слоев позволяет обучать фильтры и выделять характерные признаки букв, однако отсутствие этапов контролируемого уменьшения размерности приводит к большому числу параметров на полносвязном слое и повышает риск переобучения. Архитектура LeNet-5 является более сбалансированной, её сверточные слои извлекают признаки, пулинг уменьшает размерность и повышает устойчивость к вариациям входных данных, а полносвязная часть принимает компактное представление признаков. Это обеспечивает лучшее обобщение и приводит к максимальному качеству среди рассмотренных вариантов.

Таким образом, экспериментально подтверждено, что наличие обучаемых сверточных слоев существенно повышает точность по сравнению с использованием только пулинг-слоёв, а применение классической архитектуры LeNet-5 обеспечивает наилучшие результаты благодаря оптимальному сочетанию этапов извлечения признаков и уменьшения размерности.
